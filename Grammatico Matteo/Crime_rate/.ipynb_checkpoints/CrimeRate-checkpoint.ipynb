```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_excel(r'training.xlsx')
```

```python
print(df)
```

```python

#plt.figure(figsize=(100,80), dpi= 80)
sns.pairplot(df, vars = ["Numero reati", "Mese"], hue = "Regione", palette = "deep")
#plt.show()
```

```python
df = df.groupby("Regione").size()
df.plot(kind='pie', subplots=False, figsize=(10, 8))
plt.title("Pie Chart of crime by Region")
plt.ylabel("")
plt.show()
```
